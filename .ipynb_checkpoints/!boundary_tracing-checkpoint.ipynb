```python


from collections import deque


def mandelbrot_bondary_tracing(xarr, yarr, nmax):
    lx, ly = len(xarr), len(yarr)
    mandel = -1*np.ones((len(xarr), len(yarr)))
    done = np.zeros((lx, ly), dtype=bool)
    _queue = []
    _queue = deque()
    
    @jit
    def get(i,j):
        'Gibt das jeweilige Element, berechnet es falls noch nicht berechnet.'
        nonlocal mandel
        if mandel[i,j] == -1:
            mandel[i,j] = f(xarr[i], yarr[j], nmax)
        return mandel[i,j]
    
    @jit
    def queue(i,j):
        nonlocal _queue
        _queue.append((i,j))
        
    @jit
    def scan(i,j):
        'Berechnet das Element und die umliegenden Nachbarn, falls die nicht im Set sind, queued es sie.'
        if not 0<=i<lx or not 0<=j<ly: return
        if done[i,j]: return
        
        m = get(i,j)
        l = r = b = t = False
        if i>0:
            if mandel[i-1,j] == -1:
                if m != get(i-1,j): l=True
        if i<lx-1:
            if mandel[i+1,j] == -1:
                if m != get(i+1,j): r=True
        if j>0:
            if m != get(i,j-1): b=True
        if j<ly-1:
            if m != get(i,j+1): t=True
        if l: queue(i-1,j)
        if r: queue(i+1,j)
        if b: queue(i,j-1)
        if t: queue(i,j+1)
        if r&b: queue(i-1,j-1)
        if r&t: queue(i-1,j+1)
        if l&b: queue(i+1,j-1)
        if r&t: queue(i+1,j+1)
        done[i,j] = True
            
    @jit
    def fill():
        nonlocal mandel
        idx = mandel==-1
        mandel[idx] = nmax
    
    # Queue border points
    for i in range(lx):
        queue(i,0)
        queue(i,ly)
    for j in range(ly):
        queue(0,j)
        queue(lx,j)
        
    # Queue abarbeiten
    while len(_queue) > 0:
        #i, j = _queue[0]
        #_queue = _queue[1:]
        i, j = _queue.popleft()
        scan(i,j)
    
    # Fill inbetween pixels
    fill()
    
    return mandel
```
